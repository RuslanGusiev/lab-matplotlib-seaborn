#### Import all the libraries that are necessary.

```python
# import libraries here
import numpy as np

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

#### Define data.

```python
x = np.arange(0,100)
y = x*2
z = x**2
```

#### Plot (x,y) and (x,z) on the axes.

#### There are 2 ways of doing this. Do in both ways.

*Hint: Check out the `nrows`, `ncols`, and `index` arguments of [subplots](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.subplot.html)*

#### Also, play around with the linewidth and style. Use the ones you're most happy with.

```python
# your code here-1st way (call `subplots` twice using the `index` parameter)
plt.plot(x, y, label='double')
plt.plot(x, z, label='quadratic')

plt.xlabel('x label')
plt.ylabel('y label')

#plt.title("Double vs quadratic Plots")
#plt.legend()
#plt.show()

fig, [ax1, ax2] = plt.subplots(nrows = 1, ncols = 2, sharey = 'row') 
ax1.plot(x,y)
ax1.set_title('Linear Plot')   
ax1.set_xlabel('x')     
ax1.set_ylabel('y')   

ax2.plot(x,z)
ax2.set_title('Quadratic Plot')
ax2.set_xlabel('x')
ax2.set_ylabel('y')

```

```python
'''
L.S. First one (above) is good. Second one should have been two subplots.  
'''

# your code here-2st way (call `subplots` only once not using the `index` parameter)
plt.scatter(x,y, linewidth = 3)
plt.scatter(x,z, linewidth = 3)
plt.xlabel('X')
plt.ylabel('Y')
```

#### Augmenting your previous code, resize your previous plot.

*Hint: Add the `figsize` argument in `plt.subplots()`*

```python
# your code here
'''
L.S. Good, but the linear plot should be scaled down. 
'''

fig, [ax1, ax2] = plt.subplots(nrows = 1, ncols = 2, sharey = 'row', figsize = (15,7)) 
ax1.plot(x,y)
ax1.set_title('Linear Plot')   
ax1.set_xlabel('x')     
ax1.set_ylabel('y')   

ax2.plot(x,z)
ax2.set_title('Quadratic Plot')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
```

#### Plot both `y=x^2` and `y=exp(x)`in the same plot using normal and logarithmic scale.

*Hint: Use `set_xscale` and `set_yscale`*

```python
plt.plot(x, x**2, label='quadratic')
plt.plot(x, x**x, label='exp')

#plt.xlabel('x label')
#plt.ylabel('y label')
plt.xscale('log')
plt.yscale('log')
```

```python
'''
L.S. Here something seems to have gone wrong. I think it's due to x**x -- it would have been better to have used exp instead.  
'''

fig, [ax1, ax2] = plt.subplots(nrows = 1, ncols = 2, sharey = 'none', figsize = (15,7)) 
ax1.plot(x,x**2)
ax1.set_title('Quadratic Plot')   
ax1.set_xlabel('BlaBla')     
ax1.set_ylabel('alBalB')

ax2.plot(x, x**x)
ax2.set_title('EXP Plot')
ax2.set_xlabel('BlaBla')
ax2.set_ylabel('alBalB')
ax2.set_xscale('log')
ax2.set_yscale('log')
```
